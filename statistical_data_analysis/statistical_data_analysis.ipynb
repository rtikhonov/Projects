# Описание проекта

**Определение перспективного тарифа для телеком-компании**

Компания «Мегалайн» — федеральный оператор сотовой связи. Клиентам предлагают два тарифных плана: «Смарт» и «Ультра». Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.

Предстоит сделать предварительный анализ тарифов на небольшой выборке клиентов. В нашем распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год. Нужно проанализировать поведение клиентов и сделать вывод — какой тариф лучше.

# Описание тарифов

**Тариф «Смарт»**
- Ежемесячная плата: 550 рублей
- Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика
- Стоимость услуг сверх тарифного пакета: 1. минута разговора: 3 рубля («Мегалайн» всегда округляет вверх значения минут и мегабайтов. Если пользователь проговорил всего 1 секунду, в тарифе засчитывается целая минута); 2. сообщение: 3 рубля; 3. 1 Гб интернет-трафика: 200 рублей.

**Тариф «Ультра»**
- Ежемесячная плата: 1950 рублей
- Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика
- Стоимость услуг сверх тарифного пакета: 1. минута разговора: 1 рубль; 2. сообщение: 1 рубль; 3. 1 Гб интернет-трафика: 150 рублей.

### Объем израсходованного интернет-трафика по месяцам

```python
# добавили столбец
df_internet['month'] = df_internet['session_date'].dt.month
df_internet.head()
```

```python
# Сгруппируем, переведем данные в Гб и округлим вверх
table_internet = df_internet.groupby(['user_id', 'month']).sum(numeric_only=True)
table_internet['mb_used'] = table_internet['mb_used'] / 1024
table_internet['gb_used'] = np.ceil(table_internet['mb_used'])
table_internet.head()
```

### Помесячная выручка с каждого пользователя 

```python
def create_calculation_table(table_calls, table_messages, table_internet, df_users, df_tariffs):
    # Получаем общую таблицу
    data_table = table_calls.merge(table_messages, on=['user_id','month'], how = 'outer')\
    .merge(table_internet, on=['user_id','month'], how = 'outer')
    del data_table['mb_used']
    
    # заполним пропуски в таблице нулями
    data_table['message'].fillna(0, inplace = True)
    data_table['gb_used'].fillna(0, inplace = True)
    data_table['duration'].fillna(0, inplace = True)
    
    data_table.reset_index(inplace = True)
    
    # изменили тип на целочисленный. `duration`, 'mb_used' мы округлили вверх ранее, т.к. это требование тарифов кол-во мб тоже целочислено. 
    # Кол-во сообщений не может быть не целым.

    data_table['duration'] = data_table['duration'].astype(int)
    data_table['gb_used'] = data_table['gb_used'].astype(int)
    data_table['message'] = data_table['message'].astype(int)  
    
    
    # Присоединяем таблицу df_users
    data_table = pd.merge(data_table, df_users, on='user_id', how='inner')
    
    # Переименовываем столбец для того, чтобы можно было сделать merge 
    df_tariffs.rename(columns = {'tariff_name':"tariff"}, inplace=True)
    
    # сбросим индексы
    df_tariffs.reset_index(inplace = True)
    
    # Создаем новую таблицу для расчетов
    calculation_table = pd.merge(data_table, df_tariffs, on='tariff', how='inner')
    
    # Создаем столбцы для расчетов потраченого интернета, минут, sms
    calculation_table['internet'] = (calculation_table['mb_per_month_included'] / 1024) - calculation_table['gb_used']
    calculation_table['minutes'] = calculation_table['minutes_included'] - calculation_table['duration']
    calculation_table['messages'] = calculation_table['messages_included'] - calculation_table['message']
    
    return calculation_table

# Использование функции
calculation_table = create_calculation_table(table_calls, table_messages, table_internet, df_users, df_tariffs)
```

```python
calculation_table
```

```python
#функция для расчета переплаты за интернет. [объём интернет-трафика из тарифа - объём потраченного трафика (в мясяц)] / 1024мб * 
# * стоимость дополнительного гигабайта сверх тарифа
 
def internet_overspending(row):
    
    internet = row['internet']
    tariff = row['tariff']
    price = row['rub_per_gb']
    
    if internet < 0 and tariff == 'smart':
        return abs(internet * price)
    elif internet < 0 and tariff == 'ultra':
        return abs(internet * price)
    else:
        return 0
    
```

```python
# Создали стобец, используя функцию
calculation_table['overpayment_internet'] = calculation_table.apply(internet_overspending, axis = 1)
```

```python
calculation_table.head()
```

```python
#функция для расчета переплаты за разговоры. [количество минут разговора в месяц - длительность звонков за мясяц] * 
# * стоимость разговора сверх тарифа
def minutes_overspending(row):
    
    calls = row['minutes']
    tariff = row['tariff']
    price = row['rub_per_message']
    
    if calls < 0 and tariff == 'smart':
        return abs(calls * price)
    elif calls < 0 and tariff == 'ultra':
        return abs(calls * price)
    else:
        return 0
```

```python
calculation_table['overpayment_minutes'] = calculation_table.apply(minutes_overspending, axis = 1)
```

```python
#функция для расчета переплаты за sms. [количество sms в тарифе - кол-во отправленных сообщений в месяц] 
# * стоимость дополнительных sms

def message_overspending(row):
    
    message = row['messages']
    tariff = row['tariff']
    price = row['rub_per_minute']
    
    if message < 0 and tariff == 'smart':
        return abs(message * price)
    elif message < 0 and tariff == 'ultra':
        return abs(message * price)
    else:
        return 0
```

```python
calculation_table['overpayment_message'] = calculation_table.apply(message_overspending, axis = 1)
```

```python
# посчитаем конечную стоимость 
calculation_table['total_cost'] = calculation_table['overpayment_internet'] + calculation_table['overpayment_minutes']+\
calculation_table['overpayment_message'] + calculation_table['rub_monthly_fee']
```

### Анализ тарифов

```python
def plot_hist(df, column, title, xlabel):
    plt.figure(figsize=(10, 6))  
    sns.histplot(df[column], bins=100, color='skyblue', alpha=0.7, label=xlabel, kde=True)
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel('Клиенты', fontsize=14)
    plt.grid(True, which='both', color='black', linewidth=0.1)
    plt.legend(loc='upper right')

def plot_box(df, column, title, xlabel):
    plt.figure(figsize=(20, 2))
    plt.boxplot(df[column], vert=0)
    plt.title(title, fontsize=16)
    plt.yticks([])
    plt.xlabel(xlabel, fontsize=14)

def print_stats(df, column):
    
    maximun = df[column].max()
    minimum = df[column].min()
    mean = df[column].mean()
    var = df[column].var()
    std = df[column].std()
    
    print()
    print(200 * '*')
    print(f"Максимум: {maximun}")
    print(f"Минимум: {minimum}")
    print(f"Среднее: {mean}")
    print(f"Дисперсия: {var}")
    print(f"Стандартное откл.: {std}")
    print()
    print(200 * '*')

def analyze_tariff(final_table, tariff_name, minutes_limit, data_limit, sms_limit):
    final_table_tariff = final_table[final_table['tariff'] == tariff_name]
    print(final_table_tariff.info())
    
    # Гистограмма длительность звонка  
    plot_hist(final_table_tariff, 'duration', f'Гистограмма длительности звонков для тарифа {tariff_name}', 'Длительность звонка в минутах')
    
    # Boxplot длительность звонка
    plot_box(final_table_tariff, 'duration', f'Boxplot длительности звонков для тарифа {tariff_name}', 'Длительность звонка в минутах')
    
    # Статистика длительность звонка
    print_stats(final_table_tariff, 'duration')
    
    #Доля людей, которым не хватает минут тарифа
    print(f'Доля людей, которым не хватает {minutes_limit} минут тарифа {tariff_name} в %:',(final_table_tariff[final_table_tariff['duration'] > minutes_limit].count()[0]*100)/\
          final_table_tariff.count()[0])
    
    # Гистограмма использования интернет-трафика
    plot_hist(final_table_tariff, 'gb_used', f'Гистограмма объема потраченного за сессию интернет-трафика для тарифа {tariff_name}', 'Объём потраченного за сессию интернет-трафика')
    
    # Boxplot использования интернет-трафика
    plot_box(final_table_tariff, 'gb_used', f'Объём потраченного за сессию интернет-трафика для тарифа {tariff_name}', 'Объём потраченного за сессию интернет-трафика (в Гб)')
    
    # Статистика использования интернет-трафика
    print_stats(final_table_tariff, 'gb_used')
    
    #Доля людей, которым не хватает интернет-трафика тарифа
    print(f'Доля людей, которым не хватает {data_limit} Гб трафика тарифа для тарифа {tariff_name}',\
          (final_table_tariff[final_table_tariff['gb_used'] > data_limit].count()[0]*100)/\
          final_table_tariff.count()[0])
    
    # Гистограмма количества сообщений
    plot_hist(final_table_tariff, 'message', f'Гистограмма количества сообщений для тарифа {tariff_name}', 'Количество сообщений')
    
    # Boxplot количества сообщений
    plot_box(final_table_tariff, 'message', f'Количество сообщений для тарифа {tariff_name}', 'Количество сообщений')
    
    # Статистика количества сообщений
    print_stats(final_table_tariff, 'message')
    
    #Доля людей, которым не хватает sms тарифа
    print(f'Доля людей, которым не хватает {sms_limit} sms тарифа {tariff_name}',\
          (final_table_tariff[final_table_tariff['message'] > sms_limit].count()[0]*100)/\
          final_table_tariff.count()[0])
    
    print('Доля людей, которые не отправили ни одного sms:',\
          (final_table_tariff[final_table_tariff['message'] == 0].count()[0]*100)/\
          final_table_tariff.count()[0])
    
    print('Доля людей, которые не укладываются в тариф Smart:',\
          (final_table_tariff[final_table_tariff['total_cost'] > 550].count()[0]*100)/\
          final_table_tariff.count()[0])
    
    
    return final_table_tariff
```

### Cредняя выручка пользователей тарифов «Ультра» и «Смарт»

```python
# H0 - средняя выручка пользователей тарифов не отличаются.
# H1 - выручка отличается

ultra = final_table.query('tariff == "ultra"')['total_cost']
smart = final_table.query('tariff == "smart"')['total_cost']

alpha = .05

results = st.ttest_ind(
    ultra, 
    smart, 
    equal_var=False)

print('p-значение: ', results.pvalue) 

if results.pvalue < alpha: 
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

### Cредняя выручка пользователей из Москвы

```python
# H0  — средняя выручка пользователей из Москвы равна средней выручке пользователей из других регионов.
# H1  — средняя выручка пользователей из Москвы отличается от средней выручки пользователей из других регионов.

moscow = final_table.query('city == "Москва"')['total_cost']
other_city = final_table.query('city != "Москва"')['total_cost']

alpha = .05

results = st.ttest_ind(
    moscow, 
    other_city, 
    equal_var=False)

print('p-значение: ',results.pvalue) 

if results.pvalue < alpha: 
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

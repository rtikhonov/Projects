## Предобработка данных

```python
def preprocess_df(df):
    # Переименование столбцов
    df = df.rename({
        "cityCenters_nearest": "city_centers_nearest",
        "parks_around3000": "parks_around_3000",
        "ponds_around3000": "ponds_around_3000"
    }, axis="columns")

    # Преобразование first_day_exposition в формат даты
    df["first_day_exposition"] = pd.to_datetime(df["first_day_exposition"], format="%Y-%m-%d")
    df["first_day_exposition"] = df["first_day_exposition"].astype("datetime64[ns]")

    # Обработка аномалий в ceiling_height
    # Нулевые значения заполним модой. Модой же заполним аномально высокие и низкие потолки (более 4м и менее 2м)
    df.loc[df["ceiling_height"].isnull(), "ceiling_height"] = df["ceiling_height"].mode()[0]
    df.loc[df["ceiling_height"] > 4, "ceiling_height"] = df["ceiling_height"].mode()[0]
    df.loc[df["ceiling_height"] < 2, "ceiling_height"] = df["ceiling_height"].mode()[0]

    # Преобразование floors_total и last_price в int
    df["floors_total"] = df["floors_total"].fillna(0).astype("int")
    df["last_price"] = df["last_price"].astype("int")

    # Обработка null значений в is_apartment
    df["is_apartment"] = df["is_apartment"].fillna(False).astype("bool")

    # Обработка null значений в balcony
    # Исходили из предположения, что если не указано наличие балкона, то скорее всего его нет
    df["balcony"] = df["balcony"].fillna(0).astype("int")

    # Обработка null значений в kitchen_area для студийных квартир
    df.loc[df["studio"] == True, "kitchen_area"] = 0

    # Удаление строк с null в locality_name
    df.dropna(subset=["locality_name"], inplace=True)

    # Создание нового столбца с очищенными названиями населенных пунктов
    df["locality_name_clear"] = df["locality_name"].apply(
        lambda x: "".join([x for x in x.split() if not x[:1].islower()])
    )
    df["locality_name_clear"] = df["locality_name_clear"].str.replace("ё", "е")

    # Замена null значений в days_exposition и преобразование в int
    df["days_exposition"] = df["days_exposition"].fillna(0).astype("int")

    return df

# Вызов функции
df = preprocess_df(df)
```

## Расчёты и добавление результатов в таблицу

<div style="border:solid blue 2px; padding: 20px"> 
    
Добавим дополнительную информацию в таблицу

- цену квадратного метра;
- день недели, месяц и год публикации объявления;
- этаж квартиры; варианты — первый, последний, другой;
- соотношение жилой и общей площади, а также отношение площади кухни к общей.
    
</div>

```python
def preprocess_dataframe(df):
    # цена квадратного метра (цена на момент снятия с публикации / общая площадь)
    df["price_per_square_meter"] = df["last_price"] / df["total_area"]
    df["price_per_square_meter"] = df["price_per_square_meter"].astype("int")

    # столбец дня недели
    df["weekday"] = df["first_day_exposition"].dt.weekday

    # столбец месяца
    df["month"] = df["first_day_exposition"].dt.month

    # столбец года
    df["year"] = df["first_day_exposition"].dt.year

    # сделаем категоризацию по этажам
    def find_floor(row):
        total_floors = row["floors_total"]
        floor = row["floor"]

        if floor == 1:
            return "Первый"
        elif floor == total_floors:
            return "Последний"
        else:
            return "Другой"

    df["floor_type"] = df.apply(find_floor, axis=1)

    # соотношение жилой и общей площади
    df["living_to_total"] = df["living_area"] / df["total_area"]

    # отношение площади кухни к общей площади
    df["kitchen_to_total"] = df["kitchen_area"] / df["total_area"]

    return df

# Применяем функцию к DataFrame
df = preprocess_dataframe(df)
df.head()
```

### Изучение площади, цены, числа комнат, высоты потолков

```python
def plot_hist(df, column, bins=100):
    # Установка стиля и размера графика
    sns.set(style="whitegrid")
    plt.figure(figsize=(10, 6))

    # Построение гистограммы
    sns.histplot(df[column], bins=bins)

    # Добавление медианы, среднего значения и моды
    plt.axvline(df[column].median(), color='r', linestyle='--', label='Медиана')
    plt.axvline(df[column].mean(), color='g', linestyle='-', label='Среднее')
    plt.axvline(df[column].mode()[0], color='orange', linestyle='-', label='Мода')

    # Вычисление IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Определение границ выбросов
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Добавление выбросов на график
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)][column]
    plt.scatter(outliers, np.zeros_like(outliers), color='orange', label='Выбросы')

    # Добавление названий осей и заголовка
    plt.title(f'Распределение {column}', fontsize=16)
    plt.xlabel(column, fontsize=14)
    plt.ylabel('Частота', fontsize=14)

    # Добавление легенды
    plt.legend()

    # Отображение графика
    plt.show()

    # Вывод статистики
    print(df[column].describe())
```

```python
# Для столбца 'total_area'
plot_hist(df, 'total_area')

# Для столбца 'last_price'
plot_hist(df, 'last_price')

# Для столбца 'rooms'
plot_hist(df, 'rooms')

# Для столбца 'ceiling_height'
plot_hist(df, 'ceiling_height')
```

- есть выбросы но они не сильно влияют на данные. Основное количество комнат площадью приблизительно 50 кв. метров.
- минимальная цена ~ 12 тыс. - это подозрительно мало. Основная масса цен находится в районе 4 миллионов рублей.
- В основном комнат не больше 3. Наиболее популярны одно- и двухкомнатные квартиры. Так же есть квартиры с большим количеством комнат (5 и более).

#### Зависимость цены от даты размещения: дня недели, месяца и года

```python

def create_pivot_plot(df, index, values, aggfunc, legend, xlabel, ylabel, title):
    # Создание сводной таблицы
    pivot_table = df.pivot_table(index=[index], values=values, aggfunc=aggfunc)
    pivot_table.columns = legend

    # Построение графика
    ax = pivot_table.plot(kind='bar', grid=True, figsize=(20, 5), alpha=0.8)

    # Добавление подписей к столбцам
    for p in ax.patches:
        ax.annotate(str(round(p.get_height(), 2)), (p.get_x() * 1.005, p.get_height() * 1.005))

    # Добавление названий осей и заголовка
    plt.title(title, fontsize=16)
    plt.ylabel(ylabel, fontsize=14)
    plt.xlabel(xlabel, fontsize=14)

    # Добавление легенды
    plt.legend(legend)

    # Отображение графика
    plt.show()

    # Вывод сводной таблицы
    print(pivot_table.head(len(pivot_table)))

# Применение функции
create_pivot_plot(df_new, "weekday", "last_price", ["median", "mean"], ['Медиана', 'Среднее'], 'День недели', 'Стоимость квартиры', 'График стоимости квартиры взависимости от дня недели')
create_pivot_plot(df_new, "month", "last_price", ["median", "mean"], ['Медиана', 'Среднее'], 'Месяц', 'Стоимость квартиры', 'График стоимости квартиры в зависимости от месяца')
create_pivot_plot(df_new, "year", "last_price", ["median", "mean", "count"], ['Медиана', 'Среднее', 'Количество'], 'Год', 'Стоимость квартиры', 'График стоимости квартиры в зависимости от года')
```

## Факторы, которые влияют на стоимость квартиры

```python
import matplotlib.pyplot as plt

def create_pivot(df, index_col, values_col):
    pivot_table = df.pivot_table(
        index=[index_col], values=values_col, aggfunc=["median", "mean"]
    )

    pivot_table.columns = ["median", "mean"]
    ax = pivot_table.plot(kind='bar', 
                          grid=True, 
                          figsize=(20, 5), 
                          legend=True, 
                          linewidth=2.5, 
                          alpha=0.8, 
                          style="o")

    if index_col == "weekday":
        plt.xticks(range(7), ['Пн','Вт','Ср','Чт', 'Пт','Сб', 'Вс'])
        plt.title('График стоимости квартиры в зависимости от дня недели')
        plt.ylabel('Стоимость квартиры')
        plt.xlabel('День недели')
    elif index_col == "month":
        plt.title('График стоимости квартиры в зависимости от месяца')
        plt.ylabel('Стоимость квартиры')
        plt.xlabel('Месяц')
    elif index_col == "year":
        plt.title('График стоимости квартиры в зависимости от года')
        plt.ylabel('Стоимость квадратного метра')
        plt.xlabel('Год')

    # Добавляем легенду
    plt.legend(['Медиана', 'Среднее'], loc='upper left')

    # Добавляем статистику на график
    for p in ax.patches:
        ax.annotate(str(round(p.get_height(), 2)), (p.get_x() * 1.005, p.get_height() * 1.005))

    plt.draw()
    plt.show()

    return pivot_table.head()

```
